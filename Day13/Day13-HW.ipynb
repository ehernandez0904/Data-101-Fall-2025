Q1. The most misleading part of this graph is that the two lines “mask” and “no mask”use different y-axis. The left y-axis for mask counties ranges from 17 to 27, while the right y-axis for no-mask counties ranges from 4 to 14. This makes it look like mask counties consistently had higher case rates, even though the scales aren’t comparable. The difference between the two groups appears exaggerated, misleading viewers into thinking mask mandates increased cases, when the data might actually show the opposite or a smaller difference if plotted on the same scale.

```python
# Literally read the x, y numbers and enter them into a pandas dataframe. Then re-do the graph.


date = ['7/1/2020', '7/8/2020', '7/15/2020', '7/22/2020', '7/29/2020', 
        '8/5/2020', '8/12/2020', '8/19/2020', '8/26/2020', '9/2/2020']

# Approximate case rates for Mask (left axis)
count_mask = [26, 21, 20, 21, 22, 21, 20, 18, 17, 17]

# Approximate case rates for No-Mask (right axis)
count_nomask = [12, 11, 10, 9, 10, 9, 8, 7, 6, 5]

# Create the DataFrame
DF = pd.DataFrame({
    'date': date,
    'count_mask': count_mask,
    'count_nomask': count_nomask
})

# Display the table
DF
```

```python
fig = px.line(DF, 
              x='date', 
              y=['count_mask', 'count_nomask'], 
              title='Kansas COVID-19: Mask vs No-Mask Counties (Same Scale)',
              labels={'value': 'Cases per 100k', 'variable': 'County Type'})

fig.show()
```
