#### Initial Exploratory Data Analysis

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.io as pio
pio.renderers.defaule = 'colab'

from itables import show
```

```python
DF_raw=pd.read_csv('diabetes.csv')

DF_raw
```

As we can see above, we do have 768 observations and 9 variables to look at. This is a pretty small data set compared to others but it allows us to see only a small portion of different factors that could contribute to diabetes which could be helpful to understand as well. 

Definition of all variables:
- Pregnancies: number of times the person has been pregnant
- Glucose: plasma glucose concentration after a 2 hour test
- Blood Pressure: only looking at the diastolic blood pressure levels of each patient.
- SkinThickness: Triceps skinfold thickness (this is just a layer of fat , where excess fat could be a big contributer to type 2 diabetes because it could cause insulin resistance)
- Insulin: 2 hour serum of insulin
- BMI: body mass index (comparing weight to height)
- DiabetesPedigreeFunction: genetic predisposition score for diabetes
- age: age in years
- Outcome: 1= diabetes, 0=no diabetes
